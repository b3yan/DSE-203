# 2. Create functions and global variables
## 2.1 - Functions

```python
def calculate_mode(dataset, c1, c2, column_name):
    r = []
    for i in range(dataset.shape[0]):
        ri = dataset[column_name][i] - (dataset[column_name][i]%(c1 + c2))
        r.append(ri)
    return r
```

```python
def create_hash(r):
    h = []
    for i in range(len(r)):
        hi = hash(round(r[i]))
        h.append(hi)
    return h
```

```python
def calculate_s2(dataset, s1, s2, c1, c2, column_name):
    for j in range(len(s1)):
        if dataset[column_name][j] < s1[j] + c2:
            s2.append(s1[j] - (c1 + c2))
        else:
            s2.append(s1[j] + (c1 + c2))
```

```python
def find_candidate(dataset, candidate, candidate_id, index):
    for i in range(dataset.shape[0]):
        min_s = min(dataset['h1'][i], dataset['h2'][i])
        max_s = max(dataset['h1'][i], dataset['h2'][i])
        if min_s <= dataset['hi'][index] <= max_s:
            candidate[index].append(dataset['lat'][i])
            candidate_id[index].append(i)
    return candidate, candidate_id
```

```python
def find_satisfied_items(dataset, satisfied, candidate, candidate_id, index):     
    for item, item_id in zip(candidate[index], candidate_id[index]):
        if item >= dataset['R.A - c1'][index] and item <= dataset['R.A + c2'][index]:
            satisfied[index].append(item_id)
    return satisfied
```

```python
def join_string_list(dataset, satisfied, city_name_string_list, index):
    satisfied_city_name = []
    for k in range(len(satisfied[j])):
        satisfied_city_name.append(dataset['city'][satisfied[index][k]] + ' (' + str(dataset['lat'][satisfied[index][k]]) + ', ' + str(dataset['lng'][satisfied[index][k]]) + ')')
    city_name_string_list.append(', '.join(satisfied_city_name))
```

# 3. Truncated Hash Band Join Algorithm

## Phase 1 (building) 
### Phase 1.1 - for each tuple i of the relation R, truncate the value of its join attribute Ri.A before hashing
    ▪ ri = Ri.A – ((Ri.A) mod(c1 + c2)) /* Ri is the i-th tuple of relation R */
### Phase 1.2 – hi = hash(ri)
### Phase 1.3 – install tuple Ri in entry hi of a hash table
    ▪ All tuples falling into entry hi are chained together in a collision chain

```python
r = calculate_mode(R, c1, c2, column_name = 'lat')
r[:10]
```

```python
h = create_hash(r)
h[:10]
```

```python
df1 = pd.DataFrame(R['lat'])
df2 = pd.DataFrame(h)
df2.columns = ['hi']
f_column = df2['hi']
df1 = pd.concat([df1,f_column], axis = 1)
df1.head(10)
```

## Phase 2 (Probing)
### Phase 2.1 – Create a similar hash with Sj.B
    ▪ sj1 = Sj.B – ((Sj.B) mod(c1 + c2))
### Phase 2.2 – if Sj.B < (sj1 + c2) then sj2 = sj1 – (c1 + c2) else sj2 = sj1+ (c1 + c2)
    ▪ If a band of size c1 +  c2 of Sj.B overlaps some values of R that were truncated to the left neighbor of sj1 OR if it overlaps some values that were truncated to the right neighbor of sj1 (it cannot overlap both), we denote these neighbors as sj2
### Phase 2.3 – Create hash as before

    ▪ hj1 = hash(sj1) 

    ▪ hj2 = hash(sj2)

```python
s1 = calculate_mode(S, c1, c2, column_name = 'lat')
s1[:10]
```

```python
s2 = []
calculate_s2(S, s1, s2, c1, c2, column_name = 'lat')
s2[:10]
```

```python
h1 = create_hash(s1)
h2 = create_hash(s2)
h1[:10], h2[:10]
```

## Phase 3 (Join)
### Phase 3.1 – All tuples of R that join with Sj can be found among the tuple chains rooted at the hash table entries hj1 and hj2
### Phase 3.2 - find the candidate items with matching the following conditions
    ▪ ri is between s1 and s2
### Phase 3.3 - find satisfied matched items
    ▪ R.A – c1 ≤ S.B ≤ R.A + c2
### Phase 3.4 - join each satisfied items with formatting

```python
df3 = pd.DataFrame(R['lat'])
df4 = pd.DataFrame(h1)
df4.columns = ['h1']
f_column = df4["h1"]
df3 = pd.concat([df3,f_column], axis = 1)

df5 = pd.DataFrame(h2)
df5.columns = ['h2']
f_column = df5["h2"]
df3 = pd.concat([df3,f_column], axis = 1)

f_column = df1["hi"]
df3 = pd.concat([df3,f_column], axis = 1)

df3['R.A - c1'] = R['lat'] - c1
df3['R.A + c2'] = R['lat'] + c2
df3['city'] = R['city']
df3['lng'] = R['lng']

df3.head(10)
```

```python
candidate = {i:[] for i in range(range_length)}
candidate_id = {i:[] for i in range(range_length)}
satisfied = {i:[] for i in range(range_length)}
c_n = []

for j in range(range_length):   
    find_candidate(df3, candidate, candidate_id, index = j)
    find_satisfied_items(df3, satisfied, candidate, candidate_id, index = j)
    join_string_list(df3, satisfied, city_name_string_list = c_n, index = j)
```
